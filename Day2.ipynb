```python
import pandas as pd
df=pd.read_parquet(r"C:\Users\SBHARAN\pyspark_jobs\output\transactions.parquet")
#####################Filter — only Active transactions##############################
activeonly=df[df['status']=='Active' ]
print("Active transaction:\n",activeonly.to_string(index=False))
#####################Filter — high value transactions (amount > 10000)##############################
high_value=df[df["amount"]>10000 ]
print("\nHigh Value Transaction:\n",high_value.to_string(index=False))
#####################Filter — multiple conditions (AND)##############################
active_high_value=df[(df["status"]=="Active" ) & (df["amount"]>10000)]
#print("Active + High Value:")
print("\nHigh Value Transactions:\n", high_value.to_string(index=False))
###########Filter — multiple conditions (OR)
usd_or_gbp=df[(df["currency"]=="USD")  | (df["currency"]=="GBP")]
print("\n USD or GBP transactions: \n",usd_or_gbp.to_string(index=False))


```

Step 6 — Adding New Columns

```python
def flag_transaction(amount):
    if amount > 50000:
        return "Suscpicious"
    elif amount >10000:
        return "Review"
    else:
        return("Clear")
# Add a new column — like adding a derived port in Expression transformation    
df["flag"]=df["amount"].apply(flag_transaction)
print("With Flag column:\n",df.to_string(index=False))
# Add a tax column — 10% of amount
df["tax"]=df["amount"]*0.10
print("\nWith tax Column\n",df.to_string(index=False))
# Add a combined description column
df["description"] = df["currency"] + " - " + df["status"]
print("\nWith Description \n:",df.to_string(index=False))

# Filter active transactions
active_df = df[df["status"] == "Active"]
active_df.to_csv(r"C:\Users\SBHARAN\pyspark_jobs\output\active_transactions.csv",index=False)
print("Active Transaction Saved")
```
